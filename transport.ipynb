# Number of passengers per y length

Vehicles in a lane is limited by vehicle length and vehicle gap. For cars, vehicle gap is the stopping distance.
For trains gap is determined by control volume changes in the network but can easily be assumed to be the frequency of the service

x = vehicle length, xv = vehicle speed, sd = stopping distance, d_c = driving capacity, ppl = passengers per lane , n = number of vehicles

$$ s_d = (\frac{0.039x_v^2}{g})0.001 \qquad g = 3.4$$

For y length we solve for n

$$ n(x) + n-1(s_d) = y $$

$$ n(x+s_d) - s_d = y $$

$$ n = \frac{y+s_d}{x+s_d}$$

```python
#units are in km, 10km lane, 6 m length of minvan 16 passenger
import numpy as np
import matplotlib.pyplot as plt
Y = 10
x = 0.006
xv = np.linspace(40,240,100)
sd = (0.039*xv**2/(3.4))*0.001
d_c = (Y+sd)/(x+sd)
ppl = d_c*16
```

KEY : stopping distance and vehicle speed

```python
plt.plot(xv,sd,'-k')
plt.xlabel(r'$x_v(kmph)$', fontsize=16)
plt.ylabel(r'$s_d$(km)', fontsize=16)
```

Similar for train, except the train gap is
$$ v \times f $$

```python
#units are in km, 10km lane, 183 m length of train, 11 carriages with 1100 capacity, 5 min gap
s = 0.183
sv = 80
td = sv*(5/60)
t_c = (Y+td)/(s+td)
pps = t_c*1100
```

```python
plt.plot(xv,ppl,'-k')
plt.xlabel(r'$x_v(kmph)$', fontsize=16)
plt.ylabel(r'total passengers in lane', fontsize=16)
plt.axhline(y=pps, color='r', linestyle='-')
```

Obviously this is because stopping distance scales with velocity squared (kinectic energy). A van system can only beat trains (red line) below 70km/h at which point pph is compromised.
